```python
import numpy as np
import matplotlib.pyplot as plt
```

# Sigmoid Function

```python
# Calculate sigmoid function . 
def sigmoid(x):
    return 1/(1+np.exp(-x))

# Calculate derivative of sigmoid function . 
def derivative_sigmoid(x):
    return sigmoid(x)*(1-sigmoid(x))

# input array
x = np.array([-5,-4,-3,-2,-1,1,2,3,4,5])

prob_values = sigmoid(x)                # sigmoid fun output values 
deriv_val   = derivative_sigmoid(x)     # derivative fun output value . 
```

```python
# Note : sigmoid function takes inputs and returns a value with in a range of 0 to 1 .
plt.rcParams["figure.figsize"] = (12,4)
plt.subplot(1, 2, 1)
plt.plot(x,prob_values,marker= "o",c="b")
plt.title("sigmoid curve for given x")
plt.xticks(x)

# Note : derivative value of sigmoid function is always ranges between 0 to 0.25 ,results in vanishing gradient problem 
plt.subplot(1, 2, 2)
plt.plot(x,deriv_val,c="b")
plt.title("derivative value of sigmoid function")
plt.tight_layout()
```

# tanh Function 

```python
def tanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))

def deriv_tanh(x):
    return 1-(tanh(x)*tanh(x))

# input array
x = np.array([-5,-4,-3,-2,-1,1,2,3,4,5])
tanh_val  = tanh(x)
derv_tanh = deriv_tanh(x)
```

```python
# Note : tanh function takes inputs and returns a value with in a range of -1 to +1 .

plt.rcParams["figure.figsize"] = (12,4)
plt.subplot(1, 2, 1)
plt.plot(x,tanh_val,marker= "o",c="r")
plt.title("tanh fucntion curve for given x")
plt.xticks(x)

# Note : derivative value of tanh function is always ranges between 0 to 1 ,results in vanishing gradient problem 
plt.subplot(1, 2, 2)
plt.plot(x,derv_tanh,c="r")
plt.title("derivative value of tanh function")
plt.tight_layout()
```

# relu function (Rectified linear unit)

```python
# calculate relu function .
def relu(x):
    out = []
    for i in x:
        if i>0:
            out.append(i)
        elif(i<=0):
            out.append(0)
    return np.array(out)

# calculate relu func derivative  . 
def derv_relu(a):
    dout = []
    for i in a:
        if(i<0):
            dout.append(0)
        elif(i>0):
            dout.append(1)
    return np.array(dout)  
        
# input array
x = np.array([-5,-4,-3,-2,-1,1,2,3,4,5])
relu_out  = relu(x)
derv_relu = derv_relu(x)
```

```python
# Note : relu function is written as  f(x) = max(0,x)  . where if x<=0 then f(x)=0 , if x > 0 then f(x) = x
plt.rcParams["figure.figsize"] = (12,4)
plt.subplot(1, 2, 1)
plt.plot(x,relu_out,marker= "o",c="b")
plt.title("relu fucntion curve for given x")
plt.xticks(x)
plt.yticks(relu_out)

# Note : derivative of relu function for negative values is 0 , for positive values it is 1 .
plt.subplot(1, 2, 2)
plt.plot(x,derv_relu,c="b")
plt.xticks(x)
plt.yticks(derv_relu)
plt.title("derivative value of relu function")
plt.tight_layout()

# advantage : prevents vanishing gradient problem
# Disadvantage : Dying relu problem i.e relu returns 0 for negative inputs . 
```

# Leaky relu activation function
* similar to relu,but for negative inputs instead of returning 0 here we'll add small value to create slope to prevent dying relu  problem .

```python
# calculate relu function .
def lrelu(x):
    out = []
    for i in x:
        if i>0:
            out.append(i)
        elif(i<=0):
            out.append(0.5)
    return np.array(out)

# calculate relu func derivative  . 
def derv_lrelu(a):
    dout = []
    for i in a:
        if(i<0):
            dout.append(0.5)
        elif(i>0):
            dout.append(1)
    return np.array(dout)  
        
# input array
x = np.array([-5,-4,-3,-2,-1,1,2,3,4,5])
lrelu_out  = lrelu(x)
derv_lrelu = derv_lrelu(x)
```

```python
# Note : leakyrelu function is written as  f(x) = max(0,x)  . where if x<=0 then f(x)=small slope value , if x > 0 then f(x) = x
plt.rcParams["figure.figsize"] = (12,4)
plt.subplot(1, 2, 1)
plt.plot(x,lrelu_out,marker= "o",c="r")
plt.title("Leakyrelu fucntion curve for given x")
plt.xticks(x)
plt.yticks(lrelu_out)

# Note : derivative of leakyrelu function for negative values is small slope , for positive values it is 1 .
plt.subplot(1, 2, 2)
plt.plot(x,derv_lrelu,c="r")
plt.xticks(x)
plt.yticks(derv_lrelu)
plt.title("derivative value of Leakyrelu function")
plt.tight_layout()

# advantage : prevents vanishing gradient problem and prevents  Dying relu problem i.e relu returns 0 for negative inputs . 
```
